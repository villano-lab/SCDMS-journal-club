```python
import numpy as np
import matplotlib.pyplot as plt


# alpha = 1 keV^-1
ALPHA = 1


# Generate some random events
events = np.random.exponential(size=10000)

# I(Ei) = (1 - exp(-alpha*Ei))
I = 1 - np.exp(-ALPHA*events)

# print(I)
```

```python
# Plot events

plt.figure()
plt.hist(events, bins = 100)
# plt.savefig(f'plt_events_n_{n}')
```

```python
# Plot integral points thingy

plt.figure()
# plt.ylim(0, 1)
plt.hist(I, bins = 100)
# plt.savefig(f'plt_events_n_{n}')
```
