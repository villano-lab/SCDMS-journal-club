```python
import random

def generate_events(n: int = 10, e: int = 1) -> list:
    """
    Generate n random events within an energy range
    """
    events = []
    for i in range(n):
        events.append(
            random.uniform(0, e)
        )
    return events

n = [10, 100, 1000, 10000]
event_dict = {}
for i in n:
    event_dict[i] = generate_events(i)

print(event_dict[10])

import matplotlib.pyplot as plt

for n in event_dict.keys():
    plt.figure()
#     plt.ylim(0, 1)
    plt.scatter(event_dict[n], [0 for i in event_dict[n]], s=10)
#     plt.savefig(f'plt_events_n_{n}')
    plt.show()
```
