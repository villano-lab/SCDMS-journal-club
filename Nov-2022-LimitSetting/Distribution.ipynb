```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
di = np.random.poisson(0.5, 100)
plt.hist(di, bins = 20)
plt.yscale('log')
plt.show()
```

```python
di = np.random.uniform(0, 300, 10)
plt.hist(di, bins = 20)
plt.show()
```

```python
di = np.append(np.random.uniform(0, 9, 1000), np.random.poisson(0.5, 1000))
plt.hist(di, bins = 20)
plt.show()
```

```python
def curve (x, c):
    return np.exp(-x)+c
x = np.arange(0,10)
plt.plot(x, curve(x,0))
plt.show()
```
