```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X_10, Y_10 = np.loadtxt('/home/jovyan/Journal Club/10GeV.txt', delimiter=',', unpack=True)
X_5, Y_5 = np.loadtxt('/home/jovyan/Journal Club/5GeV.txt', delimiter=',', unpack=True)
X_2, Y_2 = np.loadtxt('/home/jovyan/Journal Club/2GeV.txt', delimiter=',', unpack=True)

plt.yscale('log')
plt.xscale('log')
plt.plot(X_10, Y_10, label = '10 GeV')
plt.plot(X_5, Y_5, label = '5 GeV')
plt.plot(X_2, Y_2, label = '2 GeV')

plt.axvline(x = 2, color = 'black', label = '2 keV Threshold')
plt.axvline(x = 0.5, color = 'red', label = '0.5 keV Threshold')

plt.title('Figure 1')
plt.xlabel('Recoil Energy [keV]')
plt.ylabel('WIMP Rate, [keV kg day]-1')

plt.legend()
 
plt.savefig('figure_1.png')

# Try to plot eq. 1.1 from the Lewin paper
```
